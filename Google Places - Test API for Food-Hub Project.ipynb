## Place Search Example

```python
query = "Food Manufacturers in Liverpool, Sydney, NSW, Australia"
```

```python
payload = {'query': query, 'key': key} #parameters that will be inserted into the URL
geocodeUrl = "https://maps.googleapis.com/maps/api/place/textsearch/json?" #Base URL for API

output = requests.get(geocodeUrl, params = payload) #Get the information from the API
output = output.json() #Turn .json oputput into Python dictionary

print("number of results: " + str(len(output['results'])))
```

```python
for i in range(0,len(output['results'])):
    pprint.pprint(output['results'][i])
```
